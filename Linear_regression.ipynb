```python
import matplotlib.pyplot as plt
# Data
hours_studied = [2, 3, 4, 5, 6]
exam_scores = [50, 55, 60, 65, 70]
# Plot
plt.scatter(hours_studied, exam_scores)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Exam Score vs Hours Studied')
plt.grid(True)
plt.show()
import numpy as np
from sklearn.linear_model import LinearRegression
# Reshape the data
X = np.array(hours_studied).reshape(-1, 1)
y = np.array(exam_scores)

# Create and fit the model
model = LinearRegression()
model.fit(X, y)
# Get the slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

print("Slope:",slope)
print("Intercept:", intercept)
# Predict the exam score for 7 hours of study
hours = 7
predicted_score = model.predict([[hours]])[0]
print(f"Predicted exam score for {hours} hours of study:", predicted_score)
# Plot
plt.scatter(hours_studied, exam_scores)
plt.plot(X, model.predict(X), color='red') # Regression line
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Exam Score vs Hours Studied')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sm

data=pd.read_csv ("/content/Salary_Data.csv")
print(data)
print(data.head())
sns.lmplot(x="YearsExperience",y="Salary",data=data)
plt.show()
X=np.array(data["YearsExperience"]).reshape(-1,1)
y=np.array(data["Salary"]).reshape(-1,1)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
regr=LinearRegression()
regr.fit(X_train,y_train)
print(regr.score(X_train,y_train))
y_pred=regr.predict(X_test)
print(y_pred)
plt.scatter(X_test,y_test,color="b")
plt.show()
plt.plot(X_test,y_pred,color="k")
plt.show()
print("R2 score",round(sm.r2_score(y_test,y_pred),2))
```
